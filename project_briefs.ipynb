## 2. 설정

```python
import pandas as pd
import os

SPREADSHEET_ID = '1RvhMG3iew-Sj8iS_vLuNmR2AZHa3VTjFmvulZht_hDQ'
SHEET_GID      = 905930239
OUTPUT_FILE    = 'project briefs.md'

# 공개 시트 CSV 다운로드 URL (인증 불필요)
CSV_URL = f'https://docs.google.com/spreadsheets/d/{SPREADSHEET_ID}/export?format=csv&gid={SHEET_GID}'

print('CSV URL:', CSV_URL)
```

## 3. 시트 데이터 읽기

```python
df = pd.read_csv(CSV_URL)

print(f'행 수: {len(df)}, 열 수: {len(df.columns)}')
print('\n헤더:')
for i, col in enumerate(df.columns):
    print(f'  [{i}] {col[:60]}')
```

## 4. 데이터 미리보기

```python
# 열 이름 단축 (표시용)
cols = df.columns.tolist()

# 주요 열 인덱스
COL_TEAM_NO   = 2   # C  팀 번호
COL_TEAM_NAME = 3   # D  팀명
COL_REPO_URL  = 4   # E  GitHub Repo URL
COL_IDEATION  = 5   # F  Ideation 내용
COL_COMMENT   = 6   # G  의견/코멘트
COL_FOR_WHOM  = 7   # H  누구를 위해
COL_PROBLEM   = 8   # I  어떤 문제의 해결
COL_TECH      = 9   # J  어떤 기술
COL_WHAT      = 10  # K  무엇 만들고자
COL_EFFECT    = 11  # L  기대 효과

preview = df.iloc[:, [COL_TEAM_NO, COL_TEAM_NAME, COL_REPO_URL, COL_FOR_WHOM, COL_WHAT]].copy()
preview.columns = ['팀번호', '팀명', 'GitHub URL', '누구를 위해', '만들고자 하는 것']
preview
```

## 5. 마크다운 생성 및 저장

```python

def md_cell(val):
    text = str(val).strip() if pd.notna(val) else ''
    if not text or text == 'nan':
        return '-'
    return text.replace('|', '\\|').replace('\n', ' ')

def md_cell_br(val):
    text = str(val).strip() if pd.notna(val) else ''
    if not text or text == 'nan':
        return '-'
    return text.replace('|', '\\|').replace('\n', '<br>')

def extract_url(val):
    text = str(val) if pd.notna(val) else ''
    for line in text.splitlines():
        line = line.strip()
        if line.startswith('http'):
            return line
    return ''

def is_empty(val):
    text = str(val).strip() if pd.notna(val) else ''
    return not text or text == 'nan'


def generate_markdown(df):
    df = df.dropna(how='all').copy()
    df.iloc[:, COL_TEAM_NO] = pd.to_numeric(df.iloc[:, COL_TEAM_NO], errors='coerce')
    df = df.sort_values(by=df.columns[COL_TEAM_NO], na_position='last')

    lines = []

    # ── 제목 ───────────────────────────────────────────
    lines.append('# 2026 Spring 전체 프로젝트 리스트\n')

    # ── 인덱스 테이블 ──────────────────────────────────
    lines.append('| 팀번호 | 팀명 | 트랙 | 프로젝트명 |')
    lines.append('|:------:|------|:----:|------------|')

    for _, row in df.iterrows():
        no     = md_cell(row.iloc[COL_TEAM_NO])
        name   = md_cell(row.iloc[COL_TEAM_NAME])
        anchor = f'team-{int(float(no))}' if no != '-' else f'team-{name}'
        lines.append(f'| [{no}](#{anchor}) | {name} |  |  |')

    lines.append('')
    lines.append('---\n')

    # ── 팀별 상세 섹션 ────────────────────────────────
    for _, row in df.iterrows():
        no     = md_cell(row.iloc[COL_TEAM_NO])
        name   = md_cell(row.iloc[COL_TEAM_NAME])
        url    = extract_url(row.iloc[COL_REPO_URL])
        anchor = f'team-{int(float(no))}' if no != '-' else f'team-{name}'

        lines.append(f'<a id="{anchor}"></a>')
        lines.append(f'## Team {no} {name}\n')

        lines.append('| 항목 | 내용 |')
        lines.append('|------|------|')

        # ── 상단 항목 (빈칸) ──
        lines.append('| 프로젝트명 |  |')
        lines.append('| 트랙 |  |')
        lines.append(f'| 팀명 | {name} |')
        lines.append('| 팀원 |  |')
        lines.append('| 지도교수 |  |')

        # ── 기존 항목 ──
        if not is_empty(row.iloc[COL_WHAT]):
            lines.append(f'| 무엇을 만들고자 하는가 | {md_cell_br(row.iloc[COL_WHAT])} |')

        if not is_empty(row.iloc[COL_FOR_WHOM]):
            lines.append(f'| 고객 (누구를 위해) | {md_cell_br(row.iloc[COL_FOR_WHOM])} |')

        if not is_empty(row.iloc[COL_PROBLEM]):
            lines.append(f'| Pain Point (해결할 문제) | {md_cell_br(row.iloc[COL_PROBLEM])} |')

        if not is_empty(row.iloc[COL_TECH]):
            lines.append(f'| 사용 기술 | {md_cell_br(row.iloc[COL_TECH])} |')

        if not is_empty(row.iloc[COL_EFFECT]):
            lines.append(f'| 기대 효과 | {md_cell_br(row.iloc[COL_EFFECT])} |')

        if url:
            lines.append(f'| GitHub Repo | [{url}]({url}) |')

        # ── 하단 항목 (빈칸) ──
        lines.append('| Team Ground Rule |  |')

        lines.append('')
        lines.append('[↑ 목록으로](#2026-spring-전체-프로젝트-리스트)\n')
        lines.append('---\n')

    return '\n'.join(lines)


# 생성 및 저장
markdown = generate_markdown(df)

output_path = os.path.join(os.getcwd(), OUTPUT_FILE)
with open(output_path, 'w', encoding='utf-8') as f:
    f.write(markdown)

print(f'저장 완료 → {output_path}')
print('\n--- 미리보기 ---')
print(markdown[:1500])
```
